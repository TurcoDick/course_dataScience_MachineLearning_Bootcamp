```python
from sklearn.datasets import make_blobs
```

```python
# isso é para gerar dados aleatórios para teste
data = make_blobs(n_samples=200, n_features=2, centers = 4,
                 cluster_std=1.8, random_state=101)
```

```python
data
```

```python
# tupla 0, pegue todas as linhas da coluna 0
# tupla 0, pegue todas as linhas da coluna 1

plt.scatter(data[0][:,0], data[0][:,1], c=data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
# rotulos previstos
kmeans.labels_
```

```python
data[1]
```

```python
# verificando se os dados previstos batem com os dados originais
fig, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1], c=kmeans.labels_,cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1], c=data[1],cmap='rainbow')
```
